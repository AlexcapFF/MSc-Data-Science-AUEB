### Power Method

```python

def PageRankPI(P, a, error):
    
    '''
    
    PageRank algorithm with Power Iteration   
    
    '''
    
    sec = time.time()
    
    n = P.shape[0]
    
    # sets an arbitrarily large number of iterations based on the estimated iterations
    k =  int((math.log10(error)/math.log10(a)))
    
    x_old = csr_matrix((np.ones((n,1))[:,0],(index[:,0],np.zeros(n))), shape = (n,1))/n # initialize x
    E = csr_matrix((np.ones((n,1))[:,0],(index[:,0],np.zeros(n))), shape = (n,1))/n 
    iters = 1
    
    #power iteration
    while iters <= k: 

        x_dot_a_vec = (a*x_old.T).dot(a_vec) 
        second_part = (x_dot_a_vec + (1-a))[0,0]* E.T #second part of the equation of the pagerank algorithm

        first_part = (a*x_old.T).dot(P) #first part of the equation of the pagerank algorithm
        
        x_new = (first_part + second_part).T # calculate x(t+1) using the formula
        #x_new = x_new/np.linalg.norm(x_new.toarray(),ord=1)
        
        if np.linalg.norm((x_old - x_new).toarray(),ord=1)<=error:  #stopping factor
            break       
            
        x_old = x_new.copy() # set x0 as x

        iters += 1

    sec2 = time.time()
    return(x_new.toarray()[:,0],iters-1, sec2-sec)
    
```

### Gauss - Seidel - With Asynchronous Solving
 

```python


def PageRankGS_Main(P, a, error):
        
    '''
    
    PageRank algorithm with Gauss-Seidel iterations including normalization and Asynchronous solving   
    
    '''
    
    sec = time.time()
    
    n = P.shape[0]
    
    # sets an arbitrarily large number of iterations based oon the estimated iterations
    k =  int((math.log10(error)/math.log10(a)))
    
    
    #-------------permutation matrix calculation--------------------
    #it is used to place the dangling nodes at the bottom of P
    per1 = np.ones((172,1))
    row1 = index[np.sum(P,axis=1) == 0]
    col1 = np.array(range(n-172,n))

    Perm1 = csr_matrix((per1[:,0],(row1,col1)), shape = (n,n)) + csr_matrix((per1[:,0],(col1, row1)), shape = (n,n))

    ones1 = np.ones((n,1))

    for i in range(col1.shape[0]):
        ones1[int(row1[i])] = 0
        ones1[int(col1[i])] = 0

    Perm1 = dia_matrix((ones1[:,0],0), shape=(n,n)) + Perm1
    # -------------------------------------------------------------    


    # ---------------calculation of L,U and P22 -------------------   
    
    P11 = (Perm1.dot(P)).dot(Perm1.T)[:n-col1.shape[0],:n-col1.shape[0]] # PAP.T

    P12 = (Perm1.dot(P)).dot(Perm1.T)[:n-col1.shape[0],n-col1.shape[0]:]
    
    IaP = (eye(n-col1.shape[0], format = 'csr') - a*P11).T
    
    L = tril(IaP, k=0, format = 'csr')
    U = triu(IaP, k=1, format = 'csr')

    # -------------------------------------------------------------
    
    b = np.zeros((n,1))
    x_old = np.zeros((n,1))
    x_new = np.zeros((n,1))
    
    b[:,0] = E_array.copy()
    b1, b2 = b[:n-col1.shape[0],0], b[n-col1.shape[0]:,0] #split the vector to nondangling and dangling sets
    
    x_old[:,0] = E_array.copy()
    x_old1, x_old2 = x_old[:n-col1.shape[0]], x_old[n-col1.shape[0]:] #split the vector to nondangling and dangling sets
    
    x_new1, x_new2 = x_new[:n-col1.shape[0]], x_new[n-col1.shape[0]:] #split the vector to nondangling and dangling sets
    
    
    iters = 1
    
    #lenear system solving
    while iters <= k:
        
        y1 = np.array(b1 - U * x_old1[:,0])
        
        x_new1[:,0] = spsolve_triangular(L, y1, lower = True) #calculate the nondangling node set
        x_new2[:,0] = a*x_new1[:,0] *P12 + b2  #calculate the dangling node set
        
        x_new = np.vstack((x_new1,x_new2)) #combine the vectors
            
        if np.linalg.norm(x_old - x_new,ord=1) <= error:  #stopping factor
            break   
            
        x_old = x_new.copy()  
        x_old1 = x_new[:n-col1.shape[0]].copy()
        x_old2 = x_new[n-col1.shape[0]:].copy()
    

        iters += 1

    #normalization outside loop
    x_new = np.vstack((x_new1,x_new2))/np.linalg.norm(np.vstack((x_new1,x_new2)), ord =1)    
    
    sec2 = time.time()
    return(x_new,iters-1, sec2-sec)
```

### Gauss - Seidel - Without Asynchronous Solving

```python
def PageRankGS_Test1(P, a, error):
    
    '''
    
    PageRank algorithm with Gauss-Seidel iterations including normalization 
    
    '''
    
    sec = time.time()
    
    n = P.shape[0]
    
    # sets an arbitrarily large number of iterations
    k =  int((math.log10(error)/math.log10(a)))
    
    #----no use of permutation matrix-----
    IaP = (eye(n, format = 'csr') - a*P).T
    
    L = tril(IaP, k=0, format = 'csr')
    U = triu(IaP, k=1, format = 'csr')

    # -------------------------------------------------------------
    
    b = np.zeros((n,1))
    x_old = np.zeros((n,1))
    x_new = np.zeros((n,1))
    
    b[:,0] = E_array.copy()
    
    x_old[:,0] = E_array.copy()
   
    
    iters = 1
    
    #linear system solving
    while iters <= k:
        
        y = np.array(b - U * x_old)
        
        x_new = spsolve_triangular(L, y, lower = True)
            
        if np.linalg.norm(x_old - x_new,ord=1) <= error:  #stopping factor
            break   
            
        x_old = x_new.copy()  

        iters += 1

   
    sec2 = time.time()
    return(x_new,iters-1, sec2-sec)
```

In fact it is quoted that they removed them and added them after the calculation of the pagerank vector. <br> Given that my implementation needed only 90 iterations and that the bias is included; it makes the values of the pagerank vector slightly larger with each iteration, which in turn leads to a faster convergence.

**Note on asynchronous solving:**<br>
In order to solve asynchronically, I created a permutation matrix that would place the rows of P which corresponded to dangling nodes, at the bottom of the matrix.<br> Then, consulting the **"Deeper Inside PageRank"** paper (page 13) by Langville and Meyer and **"A Parallel Gauss-Seidel Algorithm for Sparse Power System Matrices"** by Koester et al (page 3); I split P to P11 and P12 matrices. <br> P11 would have the non-dangling node part of P and P12 would have the necessary size to leave P11 a square matrix, without any major importance in the values it would end up with.

### Observations

My understanding is that 'a' plays the role of **statistical significance**. In statistics we choose a de-facto threhold for which we allow errors to exist in our end results. It is actually the 1-a value. <br> The lower the margin for error (a) the more precise will our statistics be to that of the real population; but, to the expense of needing more data (actually a number close to the real population). <br> However by allowing the margin for error to be larger we can "safely" perform statistical methods to our samples and confidently expect a result close to that of the real population.

Given the above remark, it is apparent that for larger a we "make" our algorithm to be more precise, which in turn requires more iterations for convergence. However by doing that we do not necessarilly get that much better results. <br> We should get results that are closer to the truth but that comes to the expense of a longer execution time, and given that we only have 281.903 pages, one can imagine the effort needed with millions of pages. <br> As a result we should prefer a lower value for a that yields close to the truth results without needing the same amount of time to run.
<br><br>
It is also stated in the **"Deeper Inside PageRank"** paper, that lower values of a would cause the algorithm to run faster but in expense of stability issues. One can assume that these stability issues would be inaccuracies in the final Ï€ vector.

```python


def PageRankGS_Main_Ex3(P, a, error):
        
    '''
    
    PageRank algorithm with Gauss-Seidel iterations including normalization and Asynchronous solving   
    
    '''
    
    sec = time.time()
    
    n = P.shape[0]
    
    # sets an arbitrarily large number of iterations based oon the estimated iterations
    k =  2
    
    
    #-------------permutation matrix calculation--------------------
    #it is used to place the dangling nodes at the bottom of P
    per1 = np.ones((172,1))
    row1 = index[np.sum(P,axis=1) == 0]
    col1 = np.array(range(n-172,n))

    Perm1 = csr_matrix((per1[:,0],(row1,col1)), shape = (n,n)) + csr_matrix((per1[:,0],(col1, row1)), shape = (n,n))

    ones1 = np.ones((n,1))

    for i in range(col1.shape[0]):
        ones1[int(row1[i])] = 0
        ones1[int(col1[i])] = 0

    Perm1 = dia_matrix((ones1[:,0],0), shape=(n,n)) + Perm1
    # -------------------------------------------------------------    


    # ---------------calculation of L,U and P22 -------------------   
    
    P11 = (Perm1.dot(P)).dot(Perm1.T)[:n-col1.shape[0],:n-col1.shape[0]] # PAP.T

    P12 = (Perm1.dot(P)).dot(Perm1.T)[:n-col1.shape[0],n-col1.shape[0]:]
    
    IaP = (eye(n-col1.shape[0], format = 'csr') - a*P11).T
    
    L = tril(IaP, k=0, format = 'csr')
    U = triu(IaP, k=1, format = 'csr')

    # -------------------------------------------------------------
    
    b = np.zeros((n,1))
    x_old = np.zeros((n,1))
    x_new = np.zeros((n,1))
    
    b[:,0] = E_array.copy()
    b1, b2 = b[:n-col1.shape[0],0], b[n-col1.shape[0]:,0] #split the vector to nondangling and dangling sets
    
    x_old[:,0] = E_array.copy()
    x_old1, x_old2 = x_old[:n-col1.shape[0]], x_old[n-col1.shape[0]:] #split the vector to nondangling and dangling sets
    
    x_new1, x_new2 = x_new[:n-col1.shape[0]], x_new[n-col1.shape[0]:] #split the vector to nondangling and dangling sets
    
    
    iters = 1
    
    #lenear system solving
    while iters <= k:
        
        y1 = np.array(b1 - U * x_old1[:,0])
        
        x_new1[:,0] = spsolve_triangular(L, y1, lower = True) #calculate the nondangling node set
        x_new2[:,0] = a*x_new1[:,0] *P12 + b2  #calculate the dangling node set
        
        x_new = np.vstack((x_new1,x_new2)) #combine the vectors
        
        #-----normalization inside loop-----
        x_new = np.vstack((x_new1,x_new2))/np.linalg.norm(np.vstack((x_new1,x_new2)), ord =1)
            
        if np.linalg.norm(x_old - x_new,ord=1) <= error:  #stopping factor
            break   
            
        x_old = x_new.copy()  
        x_old1 = x_new[:n-col1.shape[0]].copy()
        x_old2 = x_new[n-col1.shape[0]:].copy()

        iters += 1
    
    #normalization outside loop
    #x_new = np.vstack((x_new1,x_new2))/np.linalg.norm(np.vstack((x_new1,x_new2)), ord =1)
    
    sec2 = time.time()
    return(x_new,iters-1, sec2-sec)
```
